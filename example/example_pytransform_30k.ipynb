```python
import scanpy as sc
from pytransform.pytransform import pytransform
```

## Data Loading
I am using the PBMC 3K dataset from 10x genomics. This dataset consists of 3000 peripheral blood mononuclear cells (PBMCs). PBMCs concist of mononucleated circulating cells. These include T cells, B Cells, NK Cells, and  monocytes. This dataset should be small enough to run on any standard laptop. It can be obtained [here](https://www.10xgenomics.com/resources/datasets/3-k-pbm-cs-from-a-healthy-donor-1-standard-1-1-0).

```python
adata = sc.read_10x_mtx("filtered_gene_bc_matrices/hg19", cache=True)
```

```python
adata.var_names_make_unique()
```

## Preprocessing
Here we eliminate low quality cells, lowly expressed genes, etc.

```python
sc.pl.highest_expr_genes(adata, n_top=20)
```

```python
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
```

```python
adata.var['mt'] = adata.var_names.str.startswith('MT-')
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```

```python
sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'],
             jitter=0.4, multi_panel=True)
```

```python
sc.pl.scatter(adata, x='total_counts', y='pct_counts_mt')
sc.pl.scatter(adata, x='total_counts', y='n_genes_by_counts')
```

```python
adata = adata[adata.obs.n_genes_by_counts < 2500, :]
adata = adata[adata.obs.pct_counts_mt < 5, :]
```

## Normalizing
Here we compare standard count based methods to pytransform

```python
adata_logtransform = adata.copy() ## make a copy for log transforming
```

```python
sc.pp.normalize_total(adata_logtransform, target_sum=1e4)
sc.pp.log1p(adata_logtransform)
sc.pp.highly_variable_genes(adata_logtransform, min_mean=0.0125, max_mean=3, min_disp=0.5)
adata_logtransform.raw = adata_logtransform
adata_logtransform = adata_logtransform[:, adata_logtransform.var["highly_variable"]]
sc.pp.regress_out(adata_logtransform, ['total_counts', 'pct_counts_mt'])
sc.pp.scale(adata_logtransform, max_value=10)
```

```python
adata.raw = adata
```

```python
pytransform(adata, highly_variable=2000)
```

```python
sc.pp.regress_out(adata, ['pct_counts_mt'])
```

```python
sc.tl.pca(adata, svd_solver='arpack')
sc.tl.pca(adata_logtransform, svd_solver='arpack')
```

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.pp.neighbors(adata_logtransform, n_neighbors=10, n_pcs=40)
```

```python
sc.tl.umap(adata, min_dist=0.3)
sc.tl.umap(adata_logtransform, min_dist=0.3)
```

In the larger number of cells, we see a different data structure. However outliers are preserved, as well as general coloring. The logtransform implementation shrinks the scale, so colors may seem more extreme there than in our implementation. In addition, our model might still struggle with large datasets for several reasons, such as a lack of a way to logically select highly variable genes, and unsophisticated dispersion estimation.

```python
sc.pl.umap(adata, color=['CST3', 'NKG7', 'PPBP', "CD14"])
sc.pl.umap(adata_logtransform, color=['CST3', 'NKG7', 'PPBP', "CD14"])
```

```python
import numpy as np
adata.obs["log_umi"] = np.log10(adata.obs["total_counts"])
adata_logtransform.obs["log_umi"] = np.log10(adata_logtransform.obs["total_counts"])
sc.pl.umap(adata, color=["log_umi"])
sc.pl.umap(adata_logtransform, color=["log_umi"])
```
